```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")

text = "Loving the new AI features #AI #MachineLearning"
doc = nlp(text)
noun = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        noun.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)

noun_freq = Counter(noun)
verb_freq = Counter(verbs)

print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```

```python
essay_text="""
machine learning is an esssential field of artificial integlligence that focuses on the development of algorithms capable of learning from data.
Research apply statistical techiques to enable systems to imporove performance without explicit programming."""
```

```python
token_nltk = word_tokenize(essay_text)
print(token_nltk)
```

```python
pos_nltk = pos_tag(token_nltk)
print(pos_nltk)
```

```python
doc = nlp(essay_text)
pos_spacy = [(token.text, token.pos_) for token in doc]
print(pos_spacy)
```

```python
nouns_nltk = [word for word, tag in pos_nltk if tag.startswith('NN')]
verbs_nltk = [word for word, tag in pos_nltk if tag.startswith('VB')]
```

```python
nouns_spacy = [token.text for token in doc if token.pos_ == 'NOUN']
verbs_spacy = [token.text for token in doc if token.pos_ == 'VERB']
```

```python
noun_freq_nltk = Counter(nouns_nltk)
verb_freq_nltk = Counter(verbs_nltk)
print("NLTK Noun Frequency:", noun_freq_nltk)
print("NLTK Verb Frequency:", verb_freq_nltk)
```
