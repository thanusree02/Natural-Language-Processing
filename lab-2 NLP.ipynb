```python
import nltk
import spacy

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
medical_text = """
Patients with diabetes require regular monitoring of blood glucose levels.
Insulin therapy helps manage chronic conditions and prevent complications.
Early diagnosis improves patient outcomes in cardiovascular diseases.
"""
```

```python
sentences = sent_tokenize(medical_text)
print("Sentence Tokens:")
for s in sentences:
    print("-", s)
```

```python
words_nltk = word_tokenize(medical_text)
print("Word Tokens (NLTK):")
print(words_nltk)
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(medical_text)

words_spacy = [token.text for token in doc if not token.is_punct]
print("Word Tokens (spaCy):")
print(words_spacy)
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word.lower()) for word in words_nltk if word.isalpha()]

print("Stemmed Words:")
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words_nltk = [lemmatizer.lemmatize(word.lower()) for word in words_nltk if word.isalpha()]

print("Lemmatized Words (NLTK):")
print(lemmatized_words_nltk)
```

```python
lemmatized_words_spacy = [token.lemma_ for token in doc if token.is_alpha]

print("Lemmatized Words (spaCy):")
print(lemmatized_words_spacy)
```

```python
print("Stemming Output:")
print(stemmed_words)

print("\nLemmatization Output:")
print(lemmatized_words_spacy)
```
